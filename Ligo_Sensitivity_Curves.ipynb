```python
# Info

# This file shows the plot of amplitude spectral density of noise versus it's frequentz
#     measured at Livingston and Hanford.

# Reference : https://www.gw-openscience.org/events/GW150914/

# Data source L1 : https://www.gw-openscience.org/GW150914data/P1500238/L1-GDS-CALIB_STRAIN.txt
#             H1 : https://www.gw-openscience.org/GW150914data/P1500238/H1-GDS-CALIB_STRAIN.txt
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
World = '/dicos_ui_home/acye/'
mpl.rcParams['figure.dpi'] = 350
mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.axisbelow'] = True
mpl.rcParams['figure.facecolor'] = 'white'
mpl.rcParams['savefig.facecolor'] = 'white'
mpl.rcParams['axes.formatter.useoffset'] = False
```

```python
L1_Strain = np.loadtxt(World+'Data/L1-GDS-CALIB_STRAIN.txt')
H1_Strain = np.loadtxt(World+'Data/H1-GDS-CALIB_STRAIN.txt')
```

```python
x1, y1 = L1_Strain[:,0], L1_Strain[:,1]
x2, y2 = H1_Strain[:,0], H1_Strain[:,1] 
```

```python
plt.title('Sensitivity curves of aLIGO')
plt.loglog(x1,y1, label = 'L1', color = '#0015ec', linewidth=0.85)
plt.loglog(x2,y2, label = 'H1', color = '#ff8b0f', linewidth=0.85)
plt.fill_between(x1, 10**-19, y1, color='none', hatch='x', edgecolor='#0015ec')
plt.fill_between(x1, 10**-19, y2, color='none', hatch='/', edgecolor='#ff8b0f')
plt.xlabel('Frequency [Hz]')
plt.ylabel('ASD h(f)')
plt.xlim(10,5500)
plt.ylim(10**-24.5,10**-19)
plt.legend()
plt.show()
```
